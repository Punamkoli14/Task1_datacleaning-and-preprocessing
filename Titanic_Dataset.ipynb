# Feature Engineering

```python
#New column -> Family = SibSp + Parch
```

```python
data['family_size']=data['SibSp'] + data['Parch']
```

```python
data.sample(5)
```

```python
# we will engineer a new feature  by the name of famity type
# Step 1: Define the function
def family_type(size):
    if size == 1:
        return 'Alone'
    elif size <= 4:
        return 'medium'
    else:
        return 'Large'

# Step 2: Apply the function to the 'family_size' column
data['family_type'] = data['family_size'].apply(family_type)
```

```python
data.sample(5)
```

```python
#droping the SibSp ,Parch and family_size
```

```python
data.drop(columns=['SibSp' ,'Parch' , 'family_size'],inplace=True)
```

# Detecting Outliers

```python
#handing outlier in age
```

```python
data=data[data['Age']<(data['Age'].mean() +3 * data['Age'].std())]
data.shape
```

```python
#handing outliers from Fare column

Q1=np.percentile(data['Fare'],25)
Q3=np.percentile(data['Fare'],75)

outlier_low=Q1 - 1.5 * (Q3 -Q1)
outlier_high=Q3 + 1.5 * (Q3 -Q1)

data=data[(data['Fare']> outlier_low) & (data['Fare']<outlier_high)] 
```

```python
# One hot encoding

data.sample(5)

#columns to be transformed are Pclass, Sex , Embarked ,family_type

#df_encoded=pd.get_dummies(data=data, columns=['Pclass', 'Sex' , 'Embarked' ,'family_type'],drop_first=True)
df_encoded = pd.get_dummies(data=data, columns=['Pclass', 'Sex', 'Embarked', 'family_type'], drop_first=True, dtype=int)
df_encoded.head(10)
```

```python
#plt.figure(figsize=(15,6))
#sns.heatmap(df_encoded.corr(),cmap='summer')
numeric_data =df_encoded .select_dtypes(include='number')

# Plot heatmap
sns.heatmap(numeric_data.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap (Numeric Features Only)")
plt.show()
```
